```python
import tensorflow as tf
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Input, Dense, GlobalMaxPooling1D
from tensorflow.keras.models import Model
import numpy as np

# Load the dataset
dataset = pd.read_csv('preprocessed_dataset (1).csv')

# Preprocess the dataset
ingredients = dataset['Bahan']
titles = dataset['Nama']

tokenizer = Tokenizer()
tokenizer.fit_on_texts(ingredients)
encoded_ingredients = tokenizer.texts_to_sequences(ingredients)

max_seq_length = max(len(seq) for seq in encoded_ingredients)
padded_ingredients = pad_sequences(encoded_ingredients, maxlen=max_seq_length, padding='post')

num_titles = len(set(titles))
encoded_titles = pd.get_dummies(titles)

train_size = int(0.8 * len(padded_ingredients))
train_ingredients = padded_ingredients[:train_size]
train_titles = encoded_titles[:train_size]
test_ingredients = padded_ingredients[train_size:]
test_titles = encoded_titles[train_size:]

# Load pre-trained word embeddings (GloVe)
embedding_path = '/tmp/glove_embeddings/glove.6B.300d.txt'

embedding_index = {}
with open(embedding_path, encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefficients = np.asarray(values[1:], dtype='float32')
        embedding_index[word] = coefficients

embedding_dim = 300
embedding_matrix = np.zeros((len(tokenizer.word_index) + 1, embedding_dim))
for word, i in tokenizer.word_index.items():
    embedding_vector = embedding_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector

# Model architecture
input_layer = Input(shape=(max_seq_length,))
embedding_layer = tf.keras.layers.Embedding(len(tokenizer.word_index) + 1, 
                                            embedding_dim, 
                                            weights=[embedding_matrix],
                                            input_length=max_seq_length,
                                            trainable=False)(input_layer)
global_pooling_layer = GlobalMaxPooling1D()(embedding_layer)
dense_layer1 = Dense(128, activation='relu')(global_pooling_layer)
dense_layer2 = Dense(64, activation='relu')(dense_layer1)
output_layer = Dense(num_titles, activation='softmax')(dense_layer2)

model = Model(inputs=input_layer, outputs=output_layer)

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(train_ingredients, train_titles, epochs=50, batch_size=32, validation_data=(test_ingredients, test_titles))
```

```python
import pandas as pd
import re

# Function to remove emoji characters
def remove_emoji(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emoticons
                               u"\U0001F300-\U0001F5FF"  # Symbols & Pictographs
                               u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
                               u"\U0001F1E0-\U0001F1FF"  # Flags (iOS)
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"  # Enclosed Characters
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```
