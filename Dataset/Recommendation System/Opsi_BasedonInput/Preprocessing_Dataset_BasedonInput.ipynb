```python
#MENGHAPUS EMOJI PADA TEKS TWEET
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  #emoticons
        u"\U0001F300-\U0001F5FF"  #symbols & pictographs
        u"\U0001F680-\U0001F6FF"  #transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  #flags (iOS)
        u"\U00002500-\U00002BEF"  #chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"                 #dingbats
        u"\u3030"
        "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'', text)

ndf['Clean_Bahan'] = ndf['Bahan'].apply(deEmojify)
ndf
```

```python
#MEREPLACE SELURUH KETENTUAN DENGAN '' --> YANG BERARTI MENGHILANGKANNYA.
def remove2(text):
  text = re.sub(r"http\S+", '', text)     #MENGHAPUS HYPERLINK

  text = re.sub('[0-9]+', '', text)       #MENGHAPUS ANGKA

  text = re.sub(r'\$\w*', '', text)       #MENGHAPUS STOCK MARKET TICKERS --> $GE
 
  text = re.sub(r'^b"RT[\s]+', '', text)  #MENGHAPUS RT ATAU RETWEET

  text = re.sub(r"^b'[\s]+", '', text)

  text = re.sub(r"^ :[\s]+", '', text)

  text = re.sub(r'[\n]', ' ', text)       #MEREPLACE ENTER MENJADI SPASI

  text = re.sub(r'#', '',text)            #MENGHILANGKAN HASTAG

  text = re.sub(r'½', '',text) 
  text = re.sub(r'⅓', '',text) 
  text = re.sub(r'¼', '',text) 
  text = re.sub(r'¹', '',text) 
  text = re.sub(r'±', '',text) 
  text = re.sub(r'°', '',text) 
  text = re.sub(r'⏭', '',text) 

  return text

ndf['Clean_Bahan'] = ndf['Clean_Bahan'].apply(lambda x: remove2(x))
ndf.sort_values("Clean_Bahan", inplace = True)
ndf.drop_duplicates(subset ="Clean_Bahan", keep = 'first', inplace = True)   #SETELAH DATA CLEANING MAKA HAPUS DUPLIKASI DATA TWEET
ndf.reset_index(drop=True, inplace=True)
```

```python
#MENGGABUNGKAN STOPWORD MANUAL DENGAN STOPWORD PADA LIBRARY
stopword_factory = stopwords.words('indonesian')

stopword = stopword_factory + stopword_list

diction = ArrayDictionary(stopword)

str = StopWordRemover(diction)

print(stopword)
```

```python
#IMPORT SASTRAWI STEMMER
factory = StemmerFactory()
stem = factory.create_stemmer()
```

```python
#IMPORT TOKENIZE 
from nltk.tokenize import TweetTokenizer
```

```python
#MENJALANKAN TOKENISASI, STOPWORD REMOVAL, DAN STEMMING
def tweets_bersih(tweet):
    #TOKENISASI TWEET
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
    tweet_tokens = tokenizer.tokenize(tweet)
 
    tweets_clean = []    
    for word in tweet_tokens:
        if (word not in stopword):             #MENGAMBIL KATA YANG TIDAK TERDAPAT EMOTIKON (MEMASTIKAN MENGHAPUS KEMBALI EMOTIKON)
            stem_word = stem.stem(word)           #MELAKUKAN STEMMING PADA KATA
            tweets_clean.append(stem_word)
    
    return tweets_clean
     
ndf['Clean_Bahan'] = ndf['Clean_Bahan'].apply(lambda x: tweets_bersih(x))
```
